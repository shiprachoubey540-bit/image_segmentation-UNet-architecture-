```python

#    Train : 80% of benign + malignant only  (517 cases)
#    Val   : 20% of benign + malignant only  (130 cases)
#    Normal cases are EXCLUDED (as in the paper)

import os
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
from sklearn.model_selection import train_test_split
from scipy.ndimage import distance_transform_edt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau

# CONFIGURATION
IMG_HEIGHT    = 256
IMG_WIDTH     = 256
BATCH_SIZE    = 8
EPOCHS        = 90
LEARNING_RATE = 1e-4
BUSI_PATH     = r"./Dataset_BUSI_with_GT"

#  DATA LOADING (paper-exact: benign + malignant only, 80/20 split) 
def load_dataset_paper_split(dataset_path):
    """
    Replicates the paper's exact protocol:
      - Only benign and malignant cases (normal excluded)
      - 80% train, 20% val
      - Stratified by category so ratio is preserved
    """
    benign_imgs,    benign_masks    = [], []
    malignant_imgs, malignant_masks = [], []

    for category, img_list, mask_list in [
        ('benign',    benign_imgs,    benign_masks),
        ('malignant', malignant_imgs, malignant_masks),
    ]:
        cat_path = os.path.join(dataset_path, category)
        if not os.path.exists(cat_path):
            print(f"Warning: {cat_path} does not exist"); continue
        for fp in sorted(glob(os.path.join(cat_path, "*.png"))):
            if "_mask" in fp:
                continue
            mp = os.path.splitext(fp)[0] + "_mask.png"
            if os.path.exists(mp):
                img_list.append(fp)
                mask_list.append(mp)

    print(f"Benign cases   : {len(benign_imgs)}")
    print(f"Malignant cases: {len(malignant_imgs)}")

    # Stratified 80/20 split within each category (matches paper exactly)
    b_train_i, b_val_i, b_train_m, b_val_m = train_test_split(
        benign_imgs, benign_masks, test_size=0.2, random_state=42)
    m_train_i, m_val_i, m_train_m, m_val_m = train_test_split(
        malignant_imgs, malignant_masks, test_size=0.2, random_state=42)

    train_images = b_train_i + m_train_i
    train_masks  = b_train_m + m_train_m
    val_images   = b_val_i   + m_val_i
    val_masks    = b_val_m   + m_val_m

    print(f"\nTrain: {len(train_images)} (paper says 517)")
    print(f"Val  : {len(val_images)}  (paper says 130)")
    return train_images, train_masks, val_images, val_masks

#  PREPROCESSING 
def preprocess_image(image_path, target_size=(IMG_HEIGHT, IMG_WIDTH)):
    img = load_img(image_path, target_size=target_size, color_mode='grayscale')
    return img_to_array(img) / 255.0

def preprocess_mask(mask_path, target_size=(IMG_HEIGHT, IMG_WIDTH)):
    mask = load_img(mask_path, target_size=target_size, color_mode='grayscale')
    mask_array = img_to_array(mask) / 255.0
    return (mask_array > 0.5).astype(np.float32)

# ── AUGMENTATION (paper: h-flip, v-flip, Gaussian noise, affine) 
def augment(image, mask):
    if np.random.rand() > 0.5:                          # horizontal flip
        image, mask = np.fliplr(image), np.fliplr(mask)
    if np.random.rand() > 0.5:                          # vertical flip
        image, mask = np.flipud(image), np.flipud(mask)
    k = np.random.randint(0, 4)                         # random 90° rotation
    image, mask = np.rot90(image, k), np.rot90(mask, k)
    if np.random.rand() > 0.5:                          # Gaussian noise
        image = np.clip(image + np.random.normal(0, 0.02, image.shape), 0, 1)
    image = np.clip(image * np.random.uniform(0.85, 1.15), 0, 1)  # brightness
    return image, mask

def create_generator(image_paths, mask_paths, batch_size, augment_data=False):
    indices = np.arange(len(image_paths))
    while True:
        np.random.shuffle(indices)
        for start in range(0, len(image_paths), batch_size):
            batch_idx = indices[start:start + batch_size]
            imgs, masks = [], []
            for idx in batch_idx:
                img  = preprocess_image(image_paths[idx])
                mask = preprocess_mask(mask_paths[idx])
                if augment_data:
                    img, mask = augment(img, mask)
                imgs.append(img)
                masks.append(mask)
            yield np.array(imgs), np.array(masks)

# MODEL
def conv_block(inputs, num_filters, dropout_rate=0.0):
    x = layers.Conv2D(num_filters, 3, padding="same")(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.Conv2D(num_filters, 3, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    if dropout_rate > 0:
        x = layers.Dropout(dropout_rate)(x)
    return x

def encoder_block(inputs, num_filters, dropout_rate=0.0):
    x = conv_block(inputs, num_filters, dropout_rate)
    p = layers.MaxPool2D((2, 2))(x)
    return x, p

def decoder_block(inputs, skip, num_filters, dropout_rate=0.0):
    x = layers.Conv2DTranspose(num_filters, (2, 2), strides=2, padding="same")(inputs)
    x = layers.Concatenate()([x, skip])
    x = conv_block(x, num_filters, dropout_rate)
    return x

def build_unet(input_shape=(IMG_HEIGHT, IMG_WIDTH, 1)):
    inputs = layers.Input(input_shape)
    s1, p1 = encoder_block(inputs, 64,  dropout_rate=0.1)
    s2, p2 = encoder_block(p1,    128,  dropout_rate=0.1)
    s3, p3 = encoder_block(p2,    256,  dropout_rate=0.2)
    s4, p4 = encoder_block(p3,    512,  dropout_rate=0.2)
    b1     = conv_block(p4,       512,  dropout_rate=0.3)
    d1     = decoder_block(b1, s4, 512, dropout_rate=0.2)
    d2     = decoder_block(d1, s3, 256, dropout_rate=0.2)
    d3     = decoder_block(d2, s2, 128, dropout_rate=0.1)
    d4     = decoder_block(d3, s1,  64, dropout_rate=0.1)
    outputs = layers.Conv2D(1, 1, padding="same", activation="sigmoid")(d4)
    return keras.Model(inputs, outputs, name="U-Net")

#  LOSS 
def dice_coefficient(y_true, y_pred, e=1e-6):
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    return (2. * intersection) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + e)

def dice_loss(y_true, y_pred):
    return 1 - dice_coefficient(y_true, y_pred)

def combined_loss(y_true, y_pred):
    return tf.keras.losses.binary_crossentropy(y_true, y_pred) + dice_loss(y_true, y_pred)

#  ALL METRICS (matching paper's Table 1 )
def compute_all_metrics(y_true, y_pred_prob, threshold=0.5):
    y_true = y_true.squeeze().astype(np.float32)
    y_pred = (y_pred_prob.squeeze() > threshold).astype(np.float32)

    TP = np.sum(y_true * y_pred)
    FP = np.sum((1 - y_true) * y_pred)
    FN = np.sum(y_true * (1 - y_pred))
    TN = np.sum((1 - y_true) * (1 - y_pred))
    e  = 1e-6

    dice        = (2 * TP) / (2 * TP + FP + FN + e)
    jaccard     = TP / (TP + FP + FN + e)
    precision   = TP / (TP + FP + e)
    sensitivity = TP / (TP + FN + e)
    specificity = TN / (TN + FP + e)

    # ASSD & HD — skip if either mask is all-zero (matches paper's N/A)
    if y_true.sum() > 0 and y_pred.sum() > 0:
        dt_true = distance_transform_edt(1 - y_true)
        dt_pred = distance_transform_edt(1 - y_pred)
        d1 = dt_pred[y_true.astype(bool)]
        d2 = dt_true[y_pred.astype(bool)]
        all_d = np.concatenate([d1, d2])
        assd = np.mean(all_d)
        hd   = np.max(all_d)
    else:
        assd = hd = np.nan

    return dict(
        dice        = dice        * 100,
        jaccard     = jaccard     * 100,
        precision   = precision   * 100,
        sensitivity = sensitivity * 100,
        specificity = specificity * 100,
        assd        = assd,
        hd          = hd,
    )

# TRAINING 
def train_model():
    print("Loading dataset (paper split)...")
    train_images, train_masks, val_images, val_masks = load_dataset_paper_split(BUSI_PATH)

    train_gen   = create_generator(train_images, train_masks, BATCH_SIZE, augment_data=True)
    val_gen     = create_generator(val_images,   val_masks,   BATCH_SIZE, augment_data=False)
    train_steps = max(1, len(train_images) // BATCH_SIZE)
    val_steps   = max(1, len(val_images)   // BATCH_SIZE)

    print("\nBuilding U-Net...")
    model = build_unet()
    model.summary()

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=LEARNING_RATE),
        loss=combined_loss,
        metrics=[dice_coefficient]
    )

    callbacks = [
        ModelCheckpoint('best_unet_model.h5',
                        monitor='val_dice_coefficient', mode='max',
                        save_best_only=True, verbose=1),
        EarlyStopping(monitor='val_dice_coefficient', patience=15,
                      mode='max', restore_best_weights=True, verbose=1),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                          patience=5, min_lr=1e-7, verbose=1),
    ]

    print("\nTraining...")
    history = model.fit(
        train_gen, steps_per_epoch=train_steps,
        validation_data=val_gen, validation_steps=val_steps,
        epochs=EPOCHS, callbacks=callbacks
    )
    return model, history, val_images, val_masks   # val set = paper's validation set

# ── EVALUATION (on val set — same as paper's Table 1) ─────────────────────────
def evaluate_on_val(model, val_images, val_masks):
    results = {k: [] for k in ['dice','jaccard','precision','sensitivity','specificity','assd','hd']}

    for img_path, mask_path in zip(val_images, val_masks):
        img  = preprocess_image(img_path)
        mask = preprocess_mask(mask_path)
        pred = model.predict(np.expand_dims(img, axis=0), verbose=0)[0]
        m = compute_all_metrics(mask, pred)
        for k in results:
            results[k].append(m[k])

    print("\n" + "="*60)
    print(f"{'Metric':<15} {'Your Model':>12}   {'UNet (paper)':>12}")
    print("="*60)
    ref = dict(dice=70.74, jaccard=64.35, precision=74.75,
               sensitivity=75.53, specificity=97.80, assd=None, hd=None)
    for key in ['dice','jaccard','precision','sensitivity','specificity']:
        vals = [v for v in results[key] if not np.isnan(v)]
        yours = f"{np.mean(vals):.2f}%"
        paper = f"{ref[key]:.2f}%"
        print(f"{key:<15} {yours:>12}   {paper:>12}")
    for key in ['assd', 'hd']:
        vals = [v for v in results[key] if not np.isnan(v)]
        n_na = sum(1 for v in results[key] if np.isnan(v))
        yours = f"{np.mean(vals):.2f} px" if vals else "N/A"
        paper = "N/A" if key == 'assd' else "N/A"
        print(f"{key:<15} {yours:>12}   {paper:>12}   (N/A on {n_na} cases)")
    print("="*60)
    return results

#  PLOTS 
def plot_history(history):
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    axes[0].plot(history.history['loss'],     label='Train Loss')
    axes[0].plot(history.history['val_loss'], label='Val Loss')
    axes[0].set_title('Loss'); axes[0].legend(); axes[0].grid(True)
    axes[1].plot(history.history['dice_coefficient'],     label='Train Dice')
    axes[1].plot(history.history['val_dice_coefficient'], label='Val Dice')
    axes[1].set_title('Dice Coefficient'); axes[1].legend(); axes[1].grid(True)
    plt.tight_layout(); plt.show()

def visualize_predictions(model, image_paths, mask_paths, num_samples=5):
    indices = np.random.choice(len(image_paths), min(num_samples, len(image_paths)), replace=False)
    fig, axes = plt.subplots(len(indices), 3, figsize=(12, 4*len(indices)))
    for i, idx in enumerate(indices):
        img  = preprocess_image(image_paths[idx])
        mask = preprocess_mask(mask_paths[idx])
        pred = model.predict(np.expand_dims(img, axis=0), verbose=0)[0]
        axes[i,0].imshow(img.squeeze(),  cmap='gray'); axes[i,0].set_title('Image');        axes[i,0].axis('off')
        axes[i,1].imshow(mask.squeeze(), cmap='gray'); axes[i,1].set_title('Ground Truth'); axes[i,1].axis('off')
        axes[i,2].imshow(pred.squeeze(), cmap='gray'); axes[i,2].set_title('Prediction');   axes[i,2].axis('off')
    plt.tight_layout(); plt.show()

# RUN 
model, history, val_images, val_masks = train_model()
results = evaluate_on_val(model, val_images, val_masks)
plot_history(history)
visualize_predictions(model, val_images, val_masks, num_samples=5)
```
